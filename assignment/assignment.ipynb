**Q1.** To compute the sample mean of a variable $X$:
$$
m(X) = \dfrac{1}{N} \sum_{i=1}^N x_i
$$
and sample covariance of two variables $X$ and $Y$,
$$
\text{cov}(X,Y) = \dfrac{1}{N} \sum_{i=1}^N (x_i - m(X))(y_i - m(Y))).
$$
Recall, the sample variance of $X$ is
$$
s^2 = \dfrac{1}{N} \sum_{i=1}^N (x_i - m(X))^2.
$$
It can be very helpful to understand some basic properties of these statistics. If you want to write your calculations on a piece of paper, take a photo, and upload that to your GitHub repo, that's probably easiest.

1. Show that $m(a + bX) = a+b \times m(X)$.
2. Show that $\text{cov}(X,a+bY) = b \times \text{cov}(X,Y)$
3. Show that $\text{cov}(a+bX,a+bX) = b^2 \text{cov}(X,X) $, and in particular that $\text{cov}(X,X) = s^2 $.
4. Instead of the mean, consider the median. Consider transformations that are non-decreasing (if $x\ge x'$, then $g(x)\ge g(x')$), like $2+5 \times X$ or $\text{arcsinh}(X)$. Is a non-decreasing transformation of the median the median of the transformed variable? Explain. Does your answer apply to any quantile? The IQR? The range?
5. Consider a non-decreasing transformation $g()$. Is is always true that $m(g(X))= g(m(X))$?

```python
from IPython.display import display, Math

display(Math(r'm(a + bX) = \frac{1}{N}\sum_{i=1}^{N}(a + bx_i)'))
display(Math(r'= \frac{1}{N}\left(\sum_{i=1}^{N}a + b\sum_{i=1}^{N}x_i\right)'))
display(Math(r'= \frac{Na}{N} + \frac{b}{N}\sum_{i=1}^{N}x_i'))
display(Math(r'= a + b \cdot m(X) '))

display(Math(r'\mathrm{cov}(X, a+bY) = \frac{1}{N}\sum_{i=1}^{N}(x_i - m(X))\bigl((a + by_i) - (a + b\cdot m(Y))\bigr)'))
display(Math(r'= \frac{1}{N}\sum_{i=1}^{N}(x_i - m(X))\cdot b(y_i - m(Y))'))
display(Math(r'= b \cdot \frac{1}{N}\sum_{i=1}^{N}(x_i - m(X))(y_i - m(Y))'))
display(Math(r'= b \cdot \mathrm{cov}(X,Y) '))

display(Math(r'\mathrm{cov}(a+bX,\; a+bX) = b\cdot\mathrm{cov}(X,\; a+bX) = b^2\,\mathrm{cov}(X,X)'))
display(Math(r'\mathrm{cov}(X,X) = \frac{1}{N}\sum_{i=1}^{N}(x_i - m(X))^2 = s^2 '))

display(Math(r'\mathrm{Med}(g(X)) = g(\mathrm{Med}(X))'))

display(Math(r'g(m(X)) = g(2) = 4'))
display(Math(r'm(g(X)) = m(\{1, 9\}) = 5'))
display(Math(r'4 \neq 5 \implies m(g(X)) \geq g(m(X)) \quad \text{(Jensen\'s Inequality)}'))


#--------------

from IPython.display import display, Math, Markdown
display(Markdown("---"))
display(Markdown("### Question 4: Median under non-decreasing transformations"))

display(Markdown("**Claim:** If $g$ is non-decreasing, then $\\text{Med}(g(X)) = g(\\text{Med}(X))$"))

display(Markdown("**Proof:**"))
display(Math(r'\text{Let } x_{(1)} \leq x_{(2)} \leq \cdots \leq x_{(N)} \text{ be the sorted data}'))
display(Math(r'\text{Med}(X) = x_{(m)} \text{ where } m = \frac{N+1}{2} \text{ (the middle value)}'))
display(Math(r'g \text{ non-decreasing} \implies g(x_{(1)}) \leq g(x_{(2)}) \leq \cdots \leq g(x_{(N)})'))
display(Math(r'\text{The middle value of } g(x_{(1)}), \ldots, g(x_{(N)}) \text{ is } g(x_{(m)}) = g(\text{Med}(X))'))
display(Math(r'\therefore \text{Med}(g(X)) = g(\text{Med}(X)) \quad \blacksquare'))

display(Markdown("**Any quantile?** Yes:"))
display(Math(r'Q_p(g(X)) = g(Q_p(X)) \text{ for any } p \in [0,1]'))
display(Markdown("Because quantiles depend only on rank order, which $g$ preserves."))

display(Markdown("**IQR?** No:"))
display(Math(r'\text{IQR}(g(X)) = g(Q_{0.75}(X)) - g(Q_{0.25}(X)) \neq g(Q_{0.75}(X) - Q_{0.25}(X)) = g(\text{IQR}(X))'))

display(Markdown("**Range?** No:"))
display(Math(r'\text{Range}(g(X)) = g(x_{(N)}) - g(x_{(1)}) \neq g(x_{(N)} - x_{(1)}) = g(\text{Range}(X))'))

display(Markdown("---")) # I swore it use to be \hline for latex code but ig not
display(Markdown("### Question 5: Is $m(g(X)) = g(m(X))$ always true?"))

display(Markdown("**No.** This only holds for linear $g$. In general it fails."))

display(Markdown("**Counterexample:** Let $X = \\{1, 3\\}$, $g(x) = x^2$"))
display(Math(r'm(X) = \frac{1+3}{2} = 2'))
display(Math(r'g(m(X)) = g(2) = 4'))
display(Math(r'g(X) = \{1^2, 3^2\} = \{1, 9\}'))
display(Math(r'm(g(X)) = \frac{1+9}{2} = 5'))
display(Math(r'4 \neq 5 \quad \therefore m(g(X)) \neq g(m(X))'))

display(Markdown("**Why? Jensen's Inequality:** For a convex $g$:"))
display(Math(r'm(g(X)) \geq g(m(X))'))
display(Markdown("For a concave $g$:"))
display(Math(r'm(g(X)) \leq g(m(X))'))
display(Markdown("Equality holds **only** when $g$ is linear or $X$ is constant."))
```

**Q2.** This question looks at financial transfers from foreign entities to American universities. In particular, from which countries and giftors are the gifts coming from, and to which institutions are they going? 1. Load the `./data/ForeignGifts_edu.csv` dataset.

```python
# Q2.1
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
url = 'https://raw.githubusercontent.com/nickgoering/EDA/refs/heads/main/data/ForeignGifts_edu.csv'
df = pd.read_csv(url)
print(df)
print(max(df['Foreign Gift Amount']))


#Q2.2
# Histrogram of foreign gift amount
plt.hist(df['Foreign Gift Amount'], bins=100)
plt.xlabel('Foreign Gift Amount')
plt.ylabel('Frequency')
plt.show()
# The variable: Foreign Gift Amount is in regards to the financial amount per gift given. The majority are low, where some specific entitites give a large sum (Hence the left skew).
# possible data entry errors or outliers since some entries have a value of "99999999"

#Q2.3
# Histogram of Gift Type; what proportion of the gifts are contracts, real estate, and monetary?
plt.hist(df['Gift Type'], bins=100)
plt.xlabel('Gift Type')
plt.ylabel('Frequency')
plt.show()
# Total entries:
print(df['Gift Type'].value_counts())
# Contract         17274
# Monetary Gift    10936
# Real Estate         11
print(df['Gift Type'].value_counts(normalize=True))
#Proportion
#Gift Type
# Contract         0.612097
# Monetary Gift    0.387513
# Real Estate      0.000390

#Q2.4kernel density plot of the log of FGA then again but conditional on gift type, patterns?
# Kernel density plot with log x axis (Foreign Gift Amounts)
sns.kdeplot(np.log(df['Foreign Gift Amount']), shade=True)
plt.xlabel('Log of Foreign Gift Amount')
plt.ylabel('Density')
plt.show()

# Kernel density plot with log x axis (Foreign Gift Amounts) conditional on gift type
sns.kdeplot(np.log(df['Foreign Gift Amount'][df['Gift Type'] == 'Contract']), shade=True, label='Contract')
sns.kdeplot(np.log(df['Foreign Gift Amount'][df['Gift Type'] == 'Monetary Gift']), shade=True, label='Monetary Gift')
sns.kdeplot(np.log(df['Foreign Gift Amount'][df['Gift Type'] == 'Real Estate']), shade=True, label='Real Estate')
plt.xlabel('Log of Foreign Gift Amount')
plt.ylabel('Density')
plt.legend()
plt.show()
# Patterns?
# The majority of real estate donations are large in amount
# The type of gift that ranges the most in donation size is monetary gift. This is reasonable since most real estate is mor expensive (as with contracts) and a monetary gift can be any amount.

#Q2.5
# top 15 countries in terms of the number of gifts? top 15 in amount given?

#top 15 countries  in number of gifts given
print(df.groupby('Country of Giftor')['Gift Type'].count().nlargest(15))

#top 15 countries in amount given
print(df.groupby('Country of Giftor')['Foreign Gift Amount'].sum().nlargest(15))


#Q2.6
#top 15 instiutions on amount of money they receive
print(df.groupby('Institution Name')['Foreign Gift Amount'].sum().nlargest(15))
#histogram of total amount received by all institutions
plt.hist(df.groupby('Institution Name')['Foreign Gift Amount'].sum(), bins=100)
plt.xlabel('Total Amount Received by Institutions')
plt.ylabel('Frequency')
plt.show()





# Q2.7Which giftors provide the most money, in total?
print(df.groupby('Giftor Name')['Foreign Gift Amount'].sum().nlargest(15))



```

**Q4.** This question looks at a time series of the number of active oil drilling rigs in the United States over time. The data comes from the Energy Information Agency.

1. Load `./data/drilling_rigs.csv` and examine the data. How many observations? How many variables? Are numeric variables correctly read in by Pandas, or will some variables have to be typecast/coerced? Explain clearly how these data need to be cleaned.
2. To convert the `Month` variable to an ordered datetime variable, use `df['time'] = pd.to_datetime(df['Month'], format='mixed')`.
3. Let's look at `Active Well Service Rig Count (Number of Rigs)`, which is the total number of rigs over time. Make a line plot of this time series. Describe what you see.
4. Instead of levels, we want to look at change over time. Compute the first difference of  `Active Well Service Rig Count (Number of Rigs)` and plot it over time. Describe what you see.
5. The first two columns are the number of onshore and offshore rigs, respectively. Melt these columns and plot the resulting series.

```python
url='https://raw.githubusercontent.com/nickgoering/EDA/refs/heads/main/data/drilling_rigs.csv'
df_dr=pd.read_csv(url)
df_dr

#Q4.1
# How many observations, how many variables
print(df_dr.shape)
print(df_dr.columns)
# 630 observations
# 10 variables

#Are numeric variables correctly read in by pandas, or will some variables have to by typecasted/coerced?
df_dr.info()
# need to convert month to datetime variable

# Q4.2
df_dr['time'] = pd.to_datetime(df_dr['Month'], format='mixed')

# Convert 'Active Well Service Rig Count (Number of Rigs)' to numeric, coercing errors to NaN
df_dr['Active Well Service Rig Count (Number of Rigs)'] = pd.to_numeric(df_dr['Active Well Service Rig Count (Number of Rigs)'], errors='coerce')

# Verify the dtype after conversion
print(f"Dtype of 'Active Well Service Rig Count (Number of Rigs)' after conversion: {df_dr['Active Well Service Rig Count (Number of Rigs)'].dtype}")

plt.figure(figsize=(12, 6))
sns.lineplot(x='time', y='Active Well Service Rig Count (Number of Rigs)', data=df_dr)
plt.title('Active Well Service Rig Count Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Rigs')
plt.grid(True)
plt.show()

df_dr['Rig Count First Difference'] = df_dr['Active Well Service Rig Count (Number of Rigs)'].diff()

plt.figure(figsize=(12, 6))
sns.lineplot(x='time', y='Rig Count First Difference', data=df_dr)
plt.title('First Difference of Active Well Service Rig Count Over Time')
plt.xlabel('Date')
plt.ylabel('Change in Number of Rigs')
plt.grid(True)
plt.show()

df_melted = df_dr.melt(
    id_vars=['time'],
    value_vars=[
        'Crude Oil and Natural Gas Rotary Rigs in Operation, Onshore (Number of Rigs)',
        'Crude Oil and Natural Gas Rotary Rigs in Operation, Offshore (Number of Rigs)'
    ],
    var_name='Rig Type',
    value_name='Number of Rigs'
)

plt.figure(figsize=(14, 7))
sns.lineplot(x='time', y='Number of Rigs', hue='Rig Type', data=df_melted)
plt.title('Onshore vs. Offshore Rotary Rig Counts Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Rigs')
plt.grid(True)
plt.legend(title='Rig Type')
plt.show()
```
